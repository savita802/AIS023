```python
x=data.iloc[:,1:4]
y=data.iloc[:,5]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsRegressor
```

```python

Train and Test split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.30)
knn=KNeighborsRegressor(n_neighbors = 5,weights='uniform',algorithm='auto')
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error ,mean_absolute_error
import math
for i in range(1,10):
    knn=KNeighborsRegressor(n_neighbors =i)
    knn.fit(x_train, y_train)  #fit the model
    y_pred=knn.predict(x_test) #make prediction on test set
    mse = np.sqrt(mean_squared_error(y_test,y_pred)) #calculate rmse
    rmse=math.sqrt(mse) #store rmse values
    print("mae  when k is "+str(i)+"=",mse)
    
```
